```python
import pandas as pd
url = '/content/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot=True)
```

```python
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='red')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']), y=data['mpg'])
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
model.score(pd.DataFrame(data['wt']),data['mpg'])
```

```python
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black')
y_pred = model.predict(pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='green')
```

```python
model = linear_model.LinearRegression() 
X = pd.DataFrame([data['wt'], data['wt']**2]).T
y = data['mpg']
model.fit(X,y)
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('Score:', model.score(X,y))
```

```python
import numpy as np

range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=(0,))
plt.plot(range,y_pred,color='blue')
```
