```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
from sklearn import linear_model

url = '/abalone.csv'
data = pd.read_csv(url)
data.head()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot=True)
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Diameter']), y=data['Length'])
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
model.score(pd.DataFrame(data['Diameter']),data['Length'])
```

```python
data.plot(kind='scatter', x='Diameter', y='Length', figsize=(7,7), color='red')
```

```python
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Diameter']), y=data['Length'])
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('Score:', model.score(pd.DataFrame(data['Diameter']),data['Length']))

data.plot(kind='scatter', x='Diameter', y='Length', figsize=(7,7), color='black')
y_pred = model.predict(pd.DataFrame(data['Diameter']))
plt.plot(data['Diameter'],y_pred,color='red')
```

```python
model = linear_model.LinearRegression() 
X = pd.DataFrame([data['Diameter'], data['Diameter']**2]).T
y = data['Length']
model.fit(X,y)
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('Score:', model.score(X,y))

range = np.arange(0,0.8,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='Diameter', y='Length', figsize=(7,7), color='black')
plt.plot(range,y_pred,color='red')
```
