```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
from sklearn import linear_model

url = '/content/Housing.csv'
data = pd.read_csv(url)
data.head()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(data)
```

```python
data.hist()
plt.show()
```

```python
data.plot(kind='scatter', x='area', y='price', figsize=(7,7), color='blue')
plt.show()
```

```python
model = linear_model.LinearRegression() 
X = pd.DataFrame([data['area']]).T
y = data['price']
model.fit(X,y)
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('Score:', model.score(X,y))

range = np.arange(0,17000,0.1)
X_range = pd.DataFrame([range]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='area', y='price', figsize=(7,7), color='black')
plt.plot(range,y_pred,color='red')
```
